```python
list_words_corpus=[]
for i in range(0,len(df)):
    for j in df.iloc[i]['review'].split():
        list_words_corpus.append(j)
```

```python
len(list_words_corpus)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords') 
```

```python
stop_word= stopwords.words("english")
```

```python
len(stop_word)
```

```python
list_word_corpus_updated = [i for i in list_words_corpus if i.lower() not in stop_word]
```

```python
len(list_word_corpus_updated)
```

```python
k =1
dic = {}
for item in list_word_corpus_updated:
    if item in list(dic.keys()):
        dic[item]=dic[item]+1
        print(str(k)+"words processed")
        k = k+1
    else:
        dic[item]=1
        print(str(k)+"words processed")
        k = k+1
```
